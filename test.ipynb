```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
MY_API_KEY = os.getenv("GEMENI_API_KEY")
# print("MY_API_KEY:", MY_API_KEY)
genai.configure(api_key=MY_API_KEY)
```

```python
model = "models/gemini-embedding-001"  # 3072 차원을 지원하는 모델로 변경
text_to_embed = "안녕하세요! 제미나이 임베딩 모델을 테스트하고 있습니다"
```

```python
embedding_result = genai.embed_content(
    model=model,
    content=text_to_embed,
    task_type="RETRIEVAL_DOCUMENT",
    title="테스트 문서",
    output_dimensionality=3072,  # 3072 차원으로 설정
)
```

```python
print(embedding_result["embedding"])
```

```python
print(f"임베딩 차원 : {len(embedding_result['embedding'])}")
```

```python
# 사용 가능한 모델들과 그 속성 확인
for m in genai.list_models():
    if "embedding" in m.name:
        print(f"모델명: {m.name}")
        print(f"설명: {m.description}")
        print(f"지원되는 차원: {getattr(m, 'output_dimensionality', 'N/A')}")
        print("=" * 50)
```
