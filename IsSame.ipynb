```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
MY_API_KEY = os.getenv("GEMENI_API_KEY")
# print("MY_API_KEY:", MY_API_KEY)
genai.configure(api_key=MY_API_KEY)
```

```python
model = "models/gemini-embedding-001"
text_to_embed = "3072차원으로 리턴받은 앞의 796개의 숫자가 796차원으로 리턴받은 숫자가 동일한지 테스트"
```

```python
result_3072 = genai.embed_content(
    model=model,
    content=text_to_embed,
    task_type="RETRIEVAL_DOCUMENT",
    title="테스트 문서",
    output_dimensionality=3072,  # 3072 차원으로 설정
)

result_796 = genai.embed_content(
    model=model,
    content=text_to_embed,
    task_type="RETRIEVAL_DOCUMENT",
    title="테스트 문서",
    output_dimensionality=796,
)
```

```python
print(result_3072["embedding"])
print(result_796["embedding"])
```

```python
sliced_3072 = result_3072["embedding"][:796]

print(sliced_3072 == result_796["embedding"])
```

**MRL 학습으로 하나의 프로젝션 레이어로 벡터차원을 조정하면서 출력할 수 있다**
